Before we dive into the whole scraping process , I will introduce you a quite usefull function called _<sub>` ' onlynumeric() '`</sub>_.

onlynumeric is a function that :
- takes a string (sentence) as an intput 
- and keeps only the numeric place of it.

We will use this function later to this tutorial. 

```python

def onlynumeric (sentence) :
    num_list = []
    for i in sentence:
        if i.isnumeric():
            num_list.append(i)
    
    return ''.join(num_list)


# check how it works
# word = "arit1002121hme99tic"
# print(onlynumeric(word))
```
